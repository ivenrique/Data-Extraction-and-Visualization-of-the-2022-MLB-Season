# Analizaremos el comportamiento del OPS segun las posiciones de los jugadores

¿Qué significa el OPS en el béisbol?
- OPS – Promedio de Embasado más Slugging | agrega porcentaje de embasado y porcentaje de slugging para obtener un número que une a los dos.

¿Qué es el Slugging?
El slugging es una estadística para medir el poder de un bateador. 
Se calcula dividiendo el total de bases alcanzadas por turnos al bate.

#  Analisis del comportamiento del OPS x Pocisión 

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Posicion', y='OPS', data=df_fullplayer_stat)
plt.title('OPS por Posicion')
plt.xlabel('Posicion')
plt.ylabel('OPS')

plt.show()
```

El boxplot muestra la distribución estadística del OPS (On-base Plus Slugging) según las posiciones de los jugadores, lo que nos permite comparar el rendimiento de los jugadores en diferentes posiciones.
El eje x representa las diferentes posiciones de los jugadores, mientras que el eje y representa los valores de OPS.
 

Conclusiones:
De manera general el tamaño de los bigotes nos representa el amplio rango de OPS de los jugadores en cada posicion
1. La posición con el valor medio de OPS más alto es: TWP(esta categoria la tiene solo 1 jugador), pero no se puede tomar en consideracion, como valor medio  ya que es un unico jugador el de esa posicion.
2. La posición con el valor medio más bajo de OPS es: CF  con lo cual tienen menor rendimiento ofensivo
3.La posición con mayor IQR es: 3B, los jugadores de esta posicion tiene gran dispersion en su ofensiva
4.La posición con el rango intercuartil más pequeño (RIC) es el CATCHER
6. En el caso de los Outliers tenemos dos jugadores que son los que encabezaron la ofensiva de la temporada con valores considerables, inclusive son outlaiers en cualquiera de las posiciones

De manera general influyen dos facores las habilidades tecnicas y la experiencia

### Cantidad de jugadores por equipo

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df_fullplayer_stat, x="Team", bins=30)
plt.title('Histogram of Players by Team')
plt.xlabel('Team')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

Vamos a crear un dataframe con para evaluar COMO la cantidad del numero de jugadores y sus OPS pueden apuntar a los mejores equipos del campeonato

```python
# agrupa por equipo y calcula la suma de los OPS, numero de jugadores, y la media de OPS por equipo 
df_team_OPS = df_fullplayer_stat.groupby("Team")["OPS"].agg(["sum", "count", "mean"]).reset_index()

# cambio el nombre de las columnas
df_team_OPS.columns = ["Team", "Total_OPS", "Num_Jugadores", "Mean_OPS"]


df_team_OPS =df_team_OPS.sort_values(by='Mean_OPS', ascending=False)


df_team_OPS
```

Como conclusion de estas vizualizaciones en el dataframe que la mayor influencia para ser mejor equipo esta dada por la combinancion (Mayor cantidad de jugadores + Mayor OPS promedio).. esta combinacion lleva a determinar cual puede ser uno de los favoritos del campeonato
EN ESTE CASO Houston es el equipo que aporta la mayor combinacion cantidad de jugadores + mayor media OPS( EN ESE ORDEN)

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_team_OPS, x='Team', y='Mean_OPS', hue='Num_Jugadores')
plt.title('Mean OPS by Teams')
plt.xlabel('Team')
plt.ylabel('Mean OPS')
plt.xticks(rotation=45)
plt.show()
```

Jonrones Totales por equipo

```python
df_team_jonrones = df_fullplayer_stat.groupby('Team')['Jonrones'].sum().reset_index()

plt.figure(figsize=(12, 6))
sorted_df_jonrones = df_team_jonrones.sort_values(by='Jonrones')
sns.barplot(data=sorted_df_jonrones, y='Team', x='Jonrones')
plt.title('Teams vs Jonrones')
plt.xlabel('Jonrones')
plt.ylabel('Team')
plt.show()
```

En este grafico vemos como los jonrones por equipo influye la puntuacion del equipo, no obstante lo que mas aporta es el aporte colectivo, ver el caso de SD con 8 jugadores pero mean OPS por debajo de la media 0.768 la media es 0.79

A modo de conclusion, segun los datos y herramientas de este poryecto,  lo que determina mayor oportunidad de exito viene dado por el aporte(en jugadores) de cada equipo y correspondientemente que sus OPS esten por encima de la media
Lo que cumple el equipo de Los Astros de Houston...que "casualmente" se coronaron campeones de la Temporada Regular de 2022, en concordancia con los graficos y analisis realizados en el proyecto
