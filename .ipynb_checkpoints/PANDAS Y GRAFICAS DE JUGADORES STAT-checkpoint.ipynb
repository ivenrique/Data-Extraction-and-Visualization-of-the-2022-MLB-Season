```python
import numpy as np
import pandas as pd

from time import sleep
from pprint import pprint

import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df_player_stat= pd.read_csv("Jugadores_Ficha_2022_V1Nac.csv")
```

### Manipulacion del df para poder trabajar con los datos
- Las columnas de Fecha de Nac y Lugar de Nac...son string se van a castear a objetos datetine

```python
#exploracion del df
df_player_stat.info()
```

```python
# Convertir la columna 'Fecha' a tipo datetime
#los valores de fecha el mes le falta el 0, ejem(6/14/1998)..asi no se pueden convertir a datetime
#..con este bucle se añaden corrigen las fechas
lista_Nac=[]
for i in df_player_stat["Fecha Nac"]:
    if len(i)== 11:
        lista_Nac.append(i[1:])#para las fechas que tienen el mes mayor a 9 
                                #y le quita el espacio enblanco delante
    else:
        lista_Nac.append(i.replace(" ","0"))#para las fechas que tienen el mes sin cero
                                            #y le agrega el cero por delante         
    
print(lista_Nac)
```

```python
# Lo mismo pasa con la fecha debut

lista_Debut=[]
for i in df_player_stat["F_Debut"]:
    if len(i)== 11:
        lista_Debut.append(i[1:])        
    else:
        lista_Debut.append(i.replace(" ","0"))
             
print(lista_Debut)
```

```python
#convertimos los datos a objeto datetime
df_player_stat["Fecha Nac"]= pd.to_datetime(lista_Nac)
df_player_stat["F_Debut"]= pd.to_datetime(lista_Debut)
```

```python
lugar=df_player_stat["Lugar_Nacimiento"]
pais=[]
for i in lugar:
    if len(i)== 3:
        a="USA"
        pais.append(a)
    else:
        pais.append(i[1:])
pais         
```

```python
#Agregamos la columna pais
df_player_stat["Pais"]=pais
```

```python
df_player_stat["Pais"].value_counts()
```

```python
paises=df_player_stat["Pais"].value_counts().index.to_list()
paises
```

```python
#participacion de paises en los 130

plt.figure(figsize=(7,4))

sns.countplot(y=df_player_stat["Pais"], order=paises)

plt.title("Distribucion de Nacionalidades en el TOP_130 Players")

plt.show()
```

```python
#participacion de estados(USA) 

Nuevo_df=df_player_stat[df_player_stat["Pais"]=="USA"].groupby(["Lugar_Nacimiento"])["id"].count()
```

```python
Nuevo_df.sort_values(ascending= False)
```

```python
plt.figure(figsize=(7, 6))

Nuevo_df_sorted = Nuevo_df.sort_values(ascending=True)
plt.barh(Nuevo_df_sorted.index, Nuevo_df_sorted.values)

for index, value in enumerate(Nuevo_df_sorted.values):
    plt.text(value, index, str(value))

plt.title("Distribucion de estados para USA PlayersTOP_130")
plt.xlabel("Numero de Jugadores")
plt.ylabel("Estados")

plt.show()
```

## Posiciones

```python
#lista con las posiciones

descrip_posiciones = {
    "P": "Pitcher /Lanzador",
    "C": "Catcher/Receptor",
    "1B": "First Base",
    "2B": "Second Base",
    "3B": "Third Base",
    "SS": "Shortstop/Campo Corto",
    "LF": "Left Field/Jardinero Izquierdo",
    "CF": "Center Field/Jardinero Central",
    "RF": "Right Field/Jardinero Derecho",
    "DH": "Designated hitter/ Bateador designado",
    "TWP": "Two way player/Jugador bidireccional"
}

```

```python
#Resumen de posiciones de Jugadores del DataFrame
position_counts=df_player_stat["Posicion"].value_counts()
```

```python
#Grafico de distribucion de posiciones
plt.figure(figsize=(12, 6))
plt.bar(position_counts.index, position_counts.values)

plt.title("Jugadores x Posiciones en el TOP_130 Players")
plt.xlabel("Posicion")
plt.ylabel("Cantidad de Jugadores")

descrip_posiciones = {
    "P": "Pitcher /Lanzador",
    "C": "Catcher/Receptor",
    "1B": "First Base",
    "2B": "Second Base",
    "3B": "Third Base",
    "SS": "Shortstop/Campo Corto",
    "LF": "Left Field/Jardinero Izquierdo",
    "CF": "Center Field/Jardinero Central",
    "RF": "Right Field/Jardinero Derecho",
    "DH": "Designated hitter/ Bateador designado",
    "TWP": "Two way player/Jugador bidireccional"
}

leyenda = [plt.Line2D([0], [0],
                            marker='o',
                            color='w',
                            label=f'{key}: {value}',
                            markerfacecolor='blue',
                            markersize=5) for key, value in descrip_posiciones.items()]

plt.legend(handles=leyenda, loc='upper right', bbox_to_anchor=(1.4, 1.2))

for index, value in enumerate(position_counts.values):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```
